```python
# Loading iris dataset using seaborn
dataset = sns.load_dataset("iris")
dataset
```

## Line Plot

**Remember**: Line Plot always generates between two numeric values/attributes

```python
sns.lineplot(x="sepal_length", y="sepal_width", data =dataset)
plt.show()
```

### How to add titles???

```python
sns.lineplot(x="sepal_length", y="sepal_width", data =dataset)
plt.title("Iris Dataset Graph")
plt.show()
```

### How to add x limits and y limit?

```python
sns.lineplot(x="sepal_length", y="sepal_width", data =dataset)
plt.title("Iris Dataset Graph")
plt.xlim(2) #X ki limit 2 se start ho
plt.ylim(1) #Y ki limit 1 se start ho
plt.show()
```

## Setting Styles
First we need to set style and rc to none then we will set our new style

```python
sns.lineplot(x="sepal_length", y="sepal_width", data =dataset)
plt.title("Iris Dataset Graph")
sns.set_style("dark") #Styles include white,whitegrid,dark,darkgrid,ticks
sns.set_style(style=None,rc=None)
plt.show()
```

## Size of Figure
Figure size must always be define before ploting a graph

```python
plt.figure(figsize=(10,2))
sns.lineplot(x="sepal_length", y="sepal_width", data =dataset)
plt.title("Iris Dataset Graph")
plt.show()
```

### Estimators
this will work on numerical variables not on categorical data

```python
sns.barplot(x="class", y="fare",hue="sex",data=titanic,
           palette='pastel',estimator=np.mean)
plt.show()
```

### Color Saturation

```python
sns.barplot(x="class", y="fare",hue="sex",data=titanic,
           palette='pastel',estimator=np.mean,saturation=2)
plt.show()
```

## Horizontal Bar Plot
To do so just put numeric variable on the x axis

```python
sns.barplot(x="fare", y="class",hue="sex",data=titanic,
           palette='pastel',estimator=np.mean,saturation=10)
plt.show()
```

### How to give colors to individual of hue?


```python
'''To do so we have to create a dictionary which includes keys from hue individual elements and 
    color which we want to assign to each of the key. then we will assign this dict to 
    palette after hue.'''
hue_color = {'male':'black','female':'yellow'}
sns.barplot(x="fare", y="class",hue="sex",data=titanic,
           palette=hue_color)
plt.show()
```

## Giving Colors manually

```python
'''In the following code the linewidth is the width of the boundary of the bars,
    facecolor is the color inside the bars you can set them manually,
    errcolor is the color of the arrows on the bars,
    edgecolor is the of the color of the boundary of bars.
    **Note**: the range of colors is between 0(dark) and 1(white). 0.5(Gray)'''

sns.barplot(x="class",y="fare",data=titanic,
            linewidth=3,facecolor=(1,1,1,0), #Change facecolor by changing values between 0 and 1. Try 0.1,0.3,0.3,0.50
            errcolor="0",edgecolor="0") # The preferable standard size of linewidth is 3.
sns.set_style(style="white")
sns.set_style(style=None,rc=None)
plt.title("Random Colors Graph")
plt.show()
```

 ### Finding the mean in the interquartile 

```python
'''showmeans is used to mark the mean in the boxplot interquartile.
    We can set our mean marker according to our requirement
    by using meanprops and we pass dictionary to it.'''

sns.boxplot(x="survived",y="age",showmeans=True,
           meanprops={"marker":"*",
                      "markersize":"15",
                      "markeredgecolor":"white"},
           data = titanic)
plt.xlabel("How many survived",size = 14),
plt.ylabel("Agw(Years)",size = 14),
plt.title("Box plot of titanic dataset",size = 15,weight = "bold") #Commas after statement is not necessary
plt.show()
```

# Line Plot with multifacets

```python
# Loading builtin dataset of dots
dots = sns.load_dataset("dots")
dots
```

```python
#Defining a color palette
pal = sns.color_palette("rocket_r")
sns.set(style="white")
#Plotting a lineplot
sns.relplot(data=dots,x="time",y="firing_rate",
           hue="coherence",size="choice",col="align",
           kind="line",size_order=["T1","T2"],
           palette=pal,height=5,aspect=.75,facet_kws=dict(sharex=False))
```

```python
sns.set_theme(style="white")

df = sns.load_dataset("penguins")

g = sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((2200, 6800), (10, 25)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
```
